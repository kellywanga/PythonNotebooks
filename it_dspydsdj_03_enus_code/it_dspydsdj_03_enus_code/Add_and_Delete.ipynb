```python
import pandas as pd
```

#### Load the politicians DataFrame we had saved previously

```python
politicians_df = pd.read_csv('datasets/politicians.csv')

politicians_df
```

#### Create a new column
The <i>Year of Birth</i> column can be added, just as we would an entry in a Python dictionary. Here, a single value is applied to all rows

```python
politicians_df['Year of Birth'] = 1900

politicians_df
```

#### Set a list of values for the column

```python
politicians_df['Year of Birth'] = [1881, 
                                   1882, 
                                   1874, 
                                   1890, 
                                   1889, 
                                   1918, 
                                   1904]

politicians_df
```

#### Removing a column
One way to remove a column is to use the del function

```python
del politicians_df['Year of Birth']

politicians_df
```

#### Adding a new row
The loc function can be used reference a row in the DataFrame

```python
politicians_df.loc[-1] = ['Otto von Bismarck','Germany']

politicians_df
```

#### Accessing the index
The index can be referenced by the index field of a DataFrame

```python
politicians_df.index = politicians_df.index + 1

politicians_df
```

#### Sorting the DataFrame by index

```python
politicians_df = politicians_df.sort_index()

politicians_df
```

#### Deleting one row from the DataFrame

```python
politicians_df = politicians_df.drop(politicians_df.index[7])

politicians_df
```

#### Defining our own index

```python
index_values = [1815,
                1881, 
                1882, 
                1874, 
                1890, 
                1889, 
                1918]
```

```python
politicians_df.index = index_values

politicians_df
```

#### If you don't like your custom index, reset it to the default

```python
politicians_df = politicians_df.reset_index()

politicians_df
```

#### One more way to drop a column

```python
politicians_df = politicians_df.drop(columns=['index'])

politicians_df
```
