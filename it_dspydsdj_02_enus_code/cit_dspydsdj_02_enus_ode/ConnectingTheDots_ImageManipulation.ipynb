```python
import numpy as np

from scipy import ndimage
```

#### Load an image
Source: https://pixabay.com/en/koala-bear-cute-portrait-gray-1757179/

```python
img_array = ndimage.imread('datasets/koala.jpg')
```

#### The image is a 3D array
The first dimension represents the width of the image. The second dimension is it's height. These two dimensions together represent all the pixels in the image. <br />
The 3rd dimension represents the RGB values for each pixel

```python
img_array.shape
```

```python
type(img_array)
```

```python
img_array
```

#### Use Matplotlib to display the image

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img_array)

plt.show()
```

#### Slicing the image
Specify which pixels you would like to include in the sliced image

```python
sliced_image = img_array[300: , 250: , :]
```

```python
plt.imshow(sliced_image)

plt.show()
```

#### Splitting the image
Perform a split along the rows

```python
x1, x2 = np.split(img_array, 2)
```

```python
plt.imshow(x1)

plt.show()
```

```python
plt.imshow(x2)

plt.show()
```

#### Split the image along columns

```python
y1, y2 = np.split(img_array, 2, axis = 1)
```

```python
plt.imshow(y1)

plt.show()
```

```python
plt.imshow(y2)

plt.show()
```

#### Combine split images using np.concatenate()

```python
plt.imshow(np.concatenate((x1, x2)))

plt.show()
```

#### Concatenating the vertically split images 
By default, this does not re-produce the original image

```python
plt.imshow(np.concatenate((y1, y2)))

plt.show()
```

```python
plt.imshow(np.concatenate((y1, y2), axis = 1))

plt.show()
```
